```python
import pandas as pd 
import numpy as np
```

```python
#import raw data set
df = pd.read_csv(r"C:\Users\user\Downloads\Detailed Hotelinfo_v2.csv")
df.head(5)
```

```python
df.shape 
```

```python
#some data type are not numeric  (but it have to be)
df.info()
```

```python
#drop unused column
to_drop = ['Location']
df.drop(columns=to_drop, inplace=True) 
df.head()
```

```python
#rename column
df=df.rename(columns={'Unnamed: 0': 'Rank'})
```

```python
#remove all comma 
df = df.replace(',','', regex=True)
```

```python
#clear unwanted words
df["No_of_reviews"]=df["No_of_reviews"].replace("reviews", "", regex=True)
df["No_of_reviews"]=df["No_of_reviews"].replace("review", "", regex=True)
df["Hotel_class"]=df["Hotel_class"].replace("of 5 bubbles", "", regex=True)
df.head(5)
```

```python
#change to numuric value
df['MinPrice'] = pd.to_numeric(df['MinPrice'])
df['MaxPrice'] = pd.to_numeric(df['MaxPrice'])
df['Rates'] = pd.to_numeric(df['Rates'])
df['Hotel_class'] = pd.to_numeric(df['Hotel_class'])
df['No_of_reviews'] = pd.to_numeric(df['No_of_reviews'])
df.info()
```

```python
#create the column AvgPrice
df['AvgPrice'] = df.apply(
    lambda row: (row.MinPrice+row.MaxPrice)/2, axis=1)
df.head(5)
```

```python
#drop missing case
#Aince '0' implies missing value, convert them to true missing elements
df["MinPrice"]=df["MinPrice"].replace(0, np.nan)
df["MaxPrice"]=df["MaxPrice"].replace(0, np.nan)
```

```python
#20 case are droped
df=df.dropna(axis = 0, how ='any')
df.info()
```
